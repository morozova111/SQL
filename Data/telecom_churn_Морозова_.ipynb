Сводная для – "Число обращений в

```python
pd.crosstab(df['Churn'], df['Customer service calls'],margins=True)
```

```python
import plotly.express as px
fig = px.histogram(df, x= 'Customer service calls', color="Churn")
fig.show()
```

Добавим теперь в наш DataFrame бинарный признак — результат
сравнения Customer service calls > 3. И еще раз посмотрим, как он связан с
оттоком.

```python
df['Many_service_calls'] = (df['Customer service calls'] > 3).astype('int')
pd.crosstab(df['Many_service_calls'],df['Churn'],margins=True)
```

```python
import plotly.express as px
fig = px.histogram(df, x= 'Many_service_calls', color="Churn")
fig.show()
```

Отток клиентов сильно возрастает после 4-х звонков в сервисный центр.

Объединённая сводная таблица

```python
pd.crosstab(df['Many_service_calls'] & df['International plan'] , df['Churn'])
```

Исходя из вышеперечисленного, отток клиентов будет большим при подключении роуминга и более 3 звонков в сервисный центр.
